```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('static_weight2.csv')

# Calculate the number of wins
num_wins = df['win'].sum()
print(f"==>> num_wins: \n{num_wins}")

# Calculate the total number of games
total_games = df.shape[0]

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(['Wins', 'Total Games'], [num_wins, total_games], color=['blue', 'gray'])
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.title('Number of Wins vs Total Games')
plt.show()
```

```python
df = pd.read_csv('static_weight2.csv')


# Crear una columna que cuenta las victorias acumuladas
df['cumulative_win'] = df['win'].cumsum()

# Filtrar los datos para color1 igual a 1 y -1
df_color1_1 = df[df['color1'] == 1].reset_index(drop=True)
df_color1_minus1 = df[df['color1'] == -1].reset_index(drop=True)

# Calcular las victorias acumuladas para cada filtro
df_color1_1['cumulative_win'] = df_color1_1['win'].cumsum()
df_color1_minus1['cumulative_win'] = df_color1_minus1['win'].cumsum()

# Crear la gráfica de líneas
plt.figure(figsize=(10, 6))
plt.plot(df_color1_1.index, df_color1_1['cumulative_win'], label='Color1 = 1', color='blue')
plt.plot(df_color1_minus1.index, df_color1_minus1['cumulative_win'], label='Color1 = -1', color='red')
plt.xlabel('Game Index')
plt.ylabel('Cumulative Wins')
plt.title('Cumulative Wins for Color1 = 1 vs Color1 = -1')
plt.legend()
plt.grid(True)
plt.show()
```
